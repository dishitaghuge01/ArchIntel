```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_with_dqi.csv")
```

```python
def categorize_score(score):

    if score >= 8:
        return "Excellent"
    elif score >= 6:
        return "Good"
    elif score >= 4:
        return "Average"
    else:
        return "Poor"
```

```python
def evaluate_floorplan(plan_id):

    plan = df[df["plan_id"] == plan_id]

    if plan.empty:
        print("Plan not found")
        return

    dqi = plan["DQI"].values[0]

    score = dqi * 10

    category = categorize_score(score)

    print("\nFloorplan Evaluation")
    print("---------------------")

    print(f"Plan ID: {plan_id}")
    print(f"DQI: {round(dqi,3)}")
    print(f"Score: {round(score,2)} / 10")
    print(f"Category: {category}")
```

```python
evaluate_floorplan(1)
```
