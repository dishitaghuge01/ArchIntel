```python
import pandas as pd

df = pd.read_csv("floorplan_features.csv")
```

```python
df = df[df["num_rooms"] >= 2]

df = df[df["gross_area"] > 0]
df = df[df["usable_area"] > 0]
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    
    if col == "plan_id":
        continue
        
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
feature_columns = df.columns.drop("plan_id")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[feature_columns] = scaler.fit_transform(df[feature_columns])
```

```python
df.to_csv("floorplan_features_clean.csv", index=False)

print("Dataset cleaned and normalized successfully")
```
