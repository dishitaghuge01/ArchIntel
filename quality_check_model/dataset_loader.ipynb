```python
import os
from pathlib import Path
import pandas as pd
```

```python
dataset_path = Path("/Users/HP/OneDrive/Documents/mini_project/quality_check_model/cubicasa5k")
```

```python
def load_split(split_name):

    split_file = dataset_path / f"{split_name}.txt"

    with open(split_file, "r") as f:
        lines = f.read().splitlines()

    records = []

    for entry in lines:

        parts = entry.split("/")

        category = parts[0]
        plan_id = parts[1]

        svg_path = dataset_path / category / plan_id / "model.svg"

        records.append({
            "plan_id": plan_id,
            "category": category,
            "split": split_name,
            "svg_path": str(svg_path)
        })

    return records
```

```python
data = []

data.extend(load_split("train"))
data.extend(load_split("test"))
data.extend(load_split("val"))

df = pd.DataFrame(data)
```
