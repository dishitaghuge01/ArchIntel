```python
from pathlib import Path
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
```

```python
SVG_NS = {"svg": "http://www.w3.org/2000/svg"}
```

```python
def parse_svg(svg_path):

    tree = ET.parse(svg_path)
    root = tree.getroot()

    elements = {
        "paths": [],
        "polygons": [],
        "lines": [],
        "texts": []
    }

    # Extract paths
    for path in root.findall(".//svg:path", SVG_NS):
        elements["paths"].append(path.attrib)

    # Extract polygons
    for poly in root.findall(".//svg:polygon", SVG_NS):
        elements["polygons"].append(poly.attrib)

    # Extract lines
    for line in root.findall(".//svg:line", SVG_NS):
        elements["lines"].append(line.attrib)

    # Extract text labels
    for text in root.findall(".//svg:text", SVG_NS):
        elements["texts"].append({
            "text": "".join(text.itertext()),
            "attrib": text.attrib
        })

    return elements
```

```python
import pandas as pd

df = pd.read_csv("cubicasa_plan_index.csv")

print(df.head())
```

```python
print(df.iloc[0]["svg_path"])
```

```python
sample_svg = df.iloc[0]["svg_path"]

elements = parse_svg(sample_svg)

print("Paths:", len(elements["paths"]))
print("Polygons:", len(elements["polygons"]))
print("Lines:", len(elements["lines"]))
print("Texts:", len(elements["texts"]))
```

```python
parsed_data = []

for _, row in df.iterrows():

    svg_path = row["svg_path"]

    try:
        elements = parse_svg(svg_path)

        parsed_data.append({
            "plan_id": row["plan_id"],
            "elements": elements
        })

    except Exception as e:
        print("Error parsing:", svg_path)
```
